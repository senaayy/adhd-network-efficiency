```python
import os
import tarfile
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
from nilearn import datasets
from nilearn.maskers import NiftiLabelsMasker
from nilearn.connectome import ConnectivityMeasure
import warnings

# Uyarƒ± kirliliƒüini engellemek i√ßin (Resampling uyarƒ±larƒ± kafanƒ± karƒ±≈ütƒ±rmasƒ±n)
warnings.filterwarnings("ignore", category=UserWarning)

# 1. DOSYA VE ATLAS AYARLARI
# Dosyanƒ±n yeri (Senin ekran g√∂r√ºnt√ºs√ºndeki yapƒ±ya g√∂re)
tar_file = '/app/nilearn_cache/aal_SPM12/AAL3v2_for_SPM12.tar.gz'
extract_path = '/app/nilearn_cache/aal_extracted'

# Eƒüer dosya .tar uzantƒ±lƒ± g√∂r√ºn√ºyorsa (Windows bazen uzantƒ±yƒ± deƒüi≈ütirir) diye kontrol:
if not os.path.exists(tar_file):
    # Alternatif ismi dene
    tar_file = '/app/nilearn_cache/aal_SPM12/AAL3v2_for_SPM12.tar'

print(f"üì¶ Ar≈üiv kontrol ediliyor: {tar_file}")

# Ar≈üivi a√ß
if not os.path.exists(extract_path):
    try:
        # √ñnce gz olarak dene, olmazsa d√ºz tar olarak dene
        mode = "r:gz" if tar_file.endswith("gz") else "r:"
        with tarfile.open(tar_file, mode) as tar:
            tar.extractall(path=extract_path)
            print("‚úÖ Ar≈üiv ba≈üarƒ±yla a√ßƒ±ldƒ±!")
    except Exception as e:
        print(f"‚ùå Ar≈üiv a√ßma hatasƒ±: {e}")
else:
    print("‚úÖ Ar≈üiv zaten a√ßƒ±k, devam ediyoruz.")

# Atlas dosyasƒ±nƒ± bul (.nii)
atlas_filename = None
for root, dirs, files in os.walk(extract_path):
    for file in files:
        if file.endswith('.nii') and 'AAL' in file:
            atlas_filename = os.path.join(root, file)
            print(f"üß† Atlas haritasƒ±: {atlas_filename}")
            break

if not atlas_filename:
    raise FileNotFoundError("‚ùå Atlas dosyasƒ± bulunamadƒ±! L√ºtfen dosyanƒ±n indiƒüinden emin ol.")

# 2. VERƒ∞ SETƒ∞ VE MASKER
print("üì° ADHD-200 verileri y√ºkleniyor...")
adhd_dataset = datasets.fetch_adhd(n_subjects=40)

# --- KRƒ∞Tƒ∞K D√úZELTME BURADA ---
# Fenotip verisini Pandas DataFrame'e √ßeviriyoruz ki hatasƒ±z okuyalƒ±m
pheno_df = pd.DataFrame(adhd_dataset.phenotypic)
# ------------------------------

print("‚öôÔ∏è Masker hazƒ±rlanƒ±yor...")
masker = NiftiLabelsMasker(labels_img=atlas_filename, standardize='zscore_sample', 
                           detrend=True, memory='nilearn_cache', verbose=0)

efficiency_results = {'ADHD': [], 'Control': []}

# 3. ANALƒ∞Z D√ñNG√úS√ú
print("üöÄ Analiz Ba≈üladƒ±! (L√ºtfen bekleyin, i≈ülemciler √ßalƒ±≈üƒ±yor...)")

for i in range(len(adhd_dataset.func)):
    try:
        # Sinyal √áƒ±karƒ±mƒ±
        time_series = masker.fit_transform(adhd_dataset.func[i])
        
        # Konnektom
        conn = ConnectivityMeasure(kind='correlation', standardize='zscore_sample')
        matrix = conn.fit_transform([time_series])[0]
        np.fill_diagonal(matrix, 0)
        
        # Graph Theory
        threshold = np.percentile(np.abs(matrix), 80)
        G = nx.from_numpy_array(np.abs(matrix) > threshold)
        eff = nx.global_efficiency(G)
        
        # --- D√úZELTƒ∞LMƒ∞≈û TANIMLAMA ---
        # .iloc[i] kullanarak satƒ±ra g√ºvenli eri≈üim
        is_adhd = pheno_df.iloc[i]['adhd']
        label = 'ADHD' if is_adhd == 1 else 'Control'
        # -----------------------------
        
        efficiency_results[label].append(eff)
        
        if (i+1) % 5 == 0:
            print(f"‚úÖ {i+1}/40 denek ba≈üarƒ±yla i≈ülendi...")
            
    except Exception as e:
        print(f"‚ö†Ô∏è Denek {i} hatasƒ±: {e}")

# 4. SONU√áLAR
print("\nüìä ANALƒ∞Z TAMAMLANDI!")
t_stat, p_val = ttest_ind(efficiency_results['ADHD'], efficiency_results['Control'])

plt.figure(figsize=(10, 6))
plt.boxplot([efficiency_results['ADHD'], efficiency_results['Control']], 
            labels=['ADHD', 'Control'], patch_artist=True,
            boxprops=dict(facecolor='#66c2a5', color='green'),
            medianprops=dict(color='black'))
plt.title(f'ADHD vs Kontrol: Beyin Aƒü Verimliliƒüi\n(p-value: {p_val:.4f})')
plt.ylabel('Global Efficiency')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

print(f"üéØ P-Deƒüeri: {p_val:.4f}")
if p_val < 0.05:
    print("üèÜ SONU√á: Gruplar arasƒ±nda ANLAMLI bir fark bulundu!")
else:
    print("info SONU√á: Gruplar arasƒ±nda istatistiksel olarak anlamlƒ± fark bulunamadƒ±.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind

# 1. VERƒ∞LERƒ∞ HAZIRLA
adhd_data = efficiency_results['ADHD']
control_data = efficiency_results['Control']

# P-deƒüerini tekrar hesapla (garanti olsun)
t_stat, p_val = ttest_ind(adhd_data, control_data)

# 2. GRAFƒ∞K AYARLARI (Profesyonel Stil)
fig, ax = plt.subplots(figsize=(8, 6), dpi=120)

# Renk Paleti (Makale Standardƒ±)
colors = ['#FF6B6B', '#4ECDC4'] # Salmon (ADHD) ve Turkuaz (Kontrol)
labels = [f'ADHD\n(n={len(adhd_data)})', f'Saƒülƒ±klƒ± Kontrol\n(n={len(control_data)})']

# A. KUTU GRAFƒ∞ƒûƒ∞ (Box Plot)
bplot = ax.boxplot([adhd_data, control_data],
                   widths=0.5,
                   patch_artist=True,
                   labels=labels,
                   medianprops=dict(color="black", linewidth=1.5),
                   boxprops=dict(linewidth=1.5),
                   whiskerprops=dict(linewidth=1.5),
                   capprops=dict(linewidth=1.5))

# Kutularƒ± boya
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7) # Hafif ≈üeffaflƒ±k (Modern g√∂r√ºn√ºm)

# B. Bƒ∞REYSEL NOKTALAR (Jitter Plot) - BU √áOK √ñNEMLƒ∞!
# Her bir hastayƒ± nokta olarak ekliyoruz ki hoca "Verin nerede?" demesin.
# Hafif rastgelelik (noise) ekliyoruz ki noktalar √ºst √ºste binmesin.
def add_jitter(data, position, color):
    x = np.random.normal(position, 0.04, size=len(data))
    ax.scatter(x, data, alpha=0.9, color='white', edgecolors='black', s=40, zorder=10)

add_jitter(adhd_data, 1, '#333333')
add_jitter(control_data, 2, '#333333')

# C. ƒ∞STATƒ∞STƒ∞K √áƒ∞ZGƒ∞Sƒ∞ (Significance Bar)
# ƒ∞ki grubun tepesine √ßizgi √ßekip p-deƒüerini yazalƒ±m
y_max = max(max(adhd_data), max(control_data)) + 0.02
h = 0.01

ax.plot([1, 1, 2, 2], [y_max, y_max+h, y_max+h, y_max], lw=1.5, c='black')
ax.text(1.5, y_max+h+0.005, f"p = {p_val:.4f}", ha='center', va='bottom', fontsize=11, fontweight='bold')

# D. SON R√úTU≈ûLAR
ax.set_title('Global Efficiency: ADHD vs Saƒülƒ±klƒ± Kontrol', fontsize=14, fontweight='bold', pad=20)
ax.set_ylabel('Global Efficiency (Verimlilik Skoru)', fontsize=12)
ax.yaxis.grid(True, linestyle='--', alpha=0.4) # Sadece yatay √ßizgiler
ax.set_axisbelow(True) # Izgarayƒ± arkaya at

# Kenar bo≈üluklarƒ±nƒ± temizle
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```
